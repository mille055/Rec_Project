```python
import pandas as pd
import requests
from tqdm import tqdm
from bs4 import BeautifulSoup
```

```python
df_podcast = pd.read_pickle('../data/podcast_df_040423.pkl')
```

```python
df_podcast_clean = df_podcast.drop_duplicates(subset=["itunes_id"],keep="first").reset_index(drop=True)
```

```python
df_podcast_clean.head()
```

```python
def get_image(link):
    page = requests.get(link)
    soup = BeautifulSoup(page.content, 'html.parser')
    div = soup.find(attrs={'class': 'l-content-width section section--hero product-hero'})
    if div is not None:
        source = div.find_all('source')[1]
        img = source['srcset'].split()[0]
    else:
        img = None
    return img
```

```python
tqdm.pandas()
df_podcast_clean['image'] = df_podcast_clean['link'].progress_apply(lambda link: get_image(link))
```
